```python
# In which range Average price lies, what is distribution look like

sns.set(font_scale=2.0) 
from scipy.stats import norm
fig, ax = plt.subplots(figsize=(16, 8))
sns.distplot(a=avacado_df.AveragePrice, kde=False, fit=norm, color='g')

# Average Price distribution shows that for most cases price of avocado is between 1.1, 1.4
```

```python
# What are TOP 5 regions where Average price are very high?

region_list=list(avacado_df.region.unique())
average_price=[]

for i in region_list:
    x=avacado_df[avacado_df.region==i]
    region_average=sum(x.AveragePrice)/len(x)
    average_price.append(region_average)

avacado_df1=pd.DataFrame({'region_list':region_list,'average_price':average_price})
new_index=avacado_df1.average_price.sort_values(ascending=False).index.values
sorted_data=avacado_df1.reindex(new_index)

plt.figure(figsize=(24,10))
ax=sns.barplot(x=sorted_data.region_list,y=sorted_data.average_price)

plt.xticks(rotation=90)
plt.xlabel('Region')
plt.ylabel('Average Price')
plt.title('Average Price of Avocado According to Region')
```

```python
# What are TOP 5 regions where Average consumption is very high?

filter1=avacado_df.region!='TotalUS'
avacado_df1=avacado_df[filter1]

region_list=list(avacado_df1.region.unique())
average_total_volume=[]

for i in region_list:
    x=avacado_df1[avacado_df1.region==i]
    average_total_volume.append(sum(x['Total Volume'])/len(x))
df3=pd.DataFrame({'region_list':region_list,'average_total_volume':average_total_volume})

new_index=df3.average_total_volume.sort_values(ascending=False).index.values
sorted_data1=df3.reindex(new_index)

plt.figure(figsize=(22,10))
ax=sns.barplot(x=sorted_data1.region_list,y=sorted_data1.average_total_volume)

plt.xticks(rotation=90)
plt.xlabel('Region')
plt.ylabel('Average of Total Volume')
plt.title('Average of Total Volume According to Region')
```
